```python
import h5py
import re
from itertools import groupby
```

```python
def get_file_and_channel(filename):
    result = re.search(r'ch(\d+)_file(\d+)_', filename)
    file_id = int(result.group(2))
    channel_id = int(result.group(1))
    return file_id, channel_id
```

```python
muxes = []
for file in files: 
    with h5py.File(file) as f: 
        print(file)
        fid, chid = get_file_and_channel(file)
        d1 = f["/Analyses/EventDetection_000/Reads/"]
        read = [x for x in d1.keys()][0]
        mux = f["/Analyses/EventDetection_000/Reads/{0}".format(read)].attrs["start_mux"]
        muxes.append([chid, fid, mux])
```

```python
muxes = sorted(muxes, key=lambda x: x[0])
d1 = {key: len(set([x[2] for x in group])) for key, group in groupby(muxes, lambda x: x[0])}
```

```python
muxes = sorted(muxes, key=lambda x: x[2])
d2 = {key: set([x[0] for x in group if d1[x[0]] == 1]) for key, group in groupby(muxes, lambda x: x[2])}
```
