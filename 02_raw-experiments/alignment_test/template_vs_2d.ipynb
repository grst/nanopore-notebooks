```python
import editdistance
from skbio.alignment import StripedSmithWaterman
```

## mouse reads

```python
files = !find ../../../../NanoporeData/burnin_downloads/ | grep fast5
```

```python
len(files)
```

```python
reads2d = []
for file in files: 
    fasta = !poretools fasta {file} | grep -A1 --no-group-separator -e templ -e twod
    if len(fasta) != 4: continue
    reads2d.append(fasta)
```

```python
len(reads2d)
```

```python
for fasta in reads2d: 
    l = len(fasta[1])
    d = editdistance.eval(fasta[1], fasta[3])
    print(l, d, d/l)
```

```python
with open("burnin_template_2d_alignment.aln", 'w') as f:
    for fasta in reads2d: 
        query = StripedSmithWaterman(fasta[1])
        al = query(fasta[3])
        f.write(fasta[0] + "\n")
        f.write(al.aligned_query_sequence + "\n")
        f.write(al.aligned_target_sequence + "\n")
```
